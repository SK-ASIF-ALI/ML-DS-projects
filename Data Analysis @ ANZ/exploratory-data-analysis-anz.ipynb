### first_name

```python
df['first_name'].unique()
```

```python
P=df[['first_name','amount']].groupby(['first_name']).sum().sort_values(by='amount',ascending=False)[:10].reset_index()
P
```

### These are the top 10 customers who have spend huge amount in the transaction table

```python
P.plot(kind='barh', x="first_name", y="amount")
plt.title("Top 10 Spenders")
```

```python
print (df.date.min())
print (df.date.max())
```

### Year-wise Transaction Volumes

```python
date_wise_tran=df[['date','amount']].groupby(['date']).sum().sort_values(by='amount',ascending=False)[:10].reset_index()
date_wise_tran.plot(kind="bar",x="date",y="amount")
plt.title("Date Wise ")
```

### Balance

```python
Top_balance=df[['first_name','balance']].groupby(['first_name']).sum().sort_values(by='balance',ascending=False)[:10].reset_index()
Top_balance.plot(kind="bar",x="first_name",y="balance")
plt.title("Top Balancers")
```
