```r
colnames_rlog_full = colnames(rlog_lihc_full)
colnames_rlog1_full = chartr('.', '-', colnames_rlog_full)
colnames(rlog_lihc_full) = colnames_rlog1_full

all(rownames(colData_lihc_full) == colnames(rlog_lihc_full))
# Should be TRUE, otherwise wrong in previous steps codes

# View(colData_lihc_full)
```

```r
lihc_degnames_1948g_ens = read.table(file = "../input/hnsc-module-2/hnsc_deg_3000_keep_final_1948g.txt", header = TRUE, sep = "\t", row.names = 1)

lihc_degnames_1948g_ens = lihc_degnames_1948g_ens$x


rlog_lihc_deg_1948_full = rlog_lihc_full[lihc_degnames_1948g_ens,]
```

```r
# Get Gene expression with OGS Gene Names

lihc_deg_1948g_ogs_map = read.table(file = "../input/hnsc-module-2/annotLookup_tcga_deg_3000_1948g_keep.txt", sep = "\t", header = TRUE, row.names = 1)


un_ogs = unique(lihc_deg_1948g_ogs_map$hgnc_symbol)

ids = match(un_ogs, lihc_deg_1948g_ogs_map$hgnc_symbol)

```

```r
rlog_lihc_deg_1935_keep = rlog_lihc_deg_1948_full[ids,]
rownames(rlog_lihc_deg_1935_keep) = lihc_deg_1948g_ogs_map$hgnc_symbol[ids]
```

```r
which(colData_lihc_full$definition == "STN_hnsc")
stn_samples_id = which(colData_lihc_full$definition == "STN_hnsc")
lihc_samples_id = which(colData_lihc_full$definition == "PST_hnsc")

rlog_lihc_deg_1935_lihc = rlog_lihc_deg_1935_keep[,lihc_samples_id]
rlog_lihc_deg_1935_stn = rlog_lihc_deg_1935_keep[,stn_samples_id]
```

```r
rlog_lihc_deg_1935_lihc
```

```r
write.table(rlog_lihc_deg_1935_lihc, file = "./rlog_hnsc_deg_1935_pst.txt", sep = "\t", row.names = TRUE, col.names = NA)
write.table(rlog_lihc_deg_1935_stn, file = "./rlog_hnsc_deg_1935_stn.txt", sep = "\t", row.names = TRUE, col.names = NA)

```

```r
lihc_degnames_1935g_ogs = rownames(rlog_lihc_deg_1935_keep)
write.table(lihc_degnames_1935g_ogs, file = "./hnsc_degnames_1935g_ogs.txt", sep = "\t", row.names = FALSE, col.names = FALSE)


save.image("./rlog_hnsc_deg_1935g_hnsc_stn_full.RData")

```
